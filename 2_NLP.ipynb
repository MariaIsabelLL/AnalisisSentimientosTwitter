```python
from nltk import word_tokenize

tokens = word_tokenize("And now for something completely different")
print('Tokens:',tokens[:10])
```

```python
from nltk.corpus import stopwords

def removeStopword(texto):
    stopwordSpanish = stopwords.words('spanish')
    textNew= [w.lower() for w in texto if w not in stopwordSpanish]
    print(texto[:20])
    print(textNew[:20])

texto=word_tokenize("A punto de cumplirse una semana desde que Nicolás Maduro pusiera en marcha las nuevas medidas economicas para Venezuela,")
removeStopword(texto)
```

```python
raw = """Las familias suelen esconder a los niños afectados o los aíslan con los animales"""
tokens = word_tokenize(raw)
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
print("Porter Stemmer",[porter.stem(t) for t in tokens])
print("Lancaster Stemmer",[lancaster.stem(t) for t in tokens])

wnl = nltk.WordNetLemmatizer()
print("Lemmatization",[wnl.lemmatize(t) for t in tokens])
```
